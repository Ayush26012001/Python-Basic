# Shift Operator
- Shift the bits by left or right
1. Left Shift (<<) - Shift the bits to the left by a specified no of position, filling 0's of the right.
2. Right Shift (>>)- remove the number of element in the binary  specified no of position, removings 0's from the right.

```python
35<<3
```

```python
bin(35)
```

```python
5<<1
```

```python
35>>3
```

```python
280>>3
```

```python
10>>1
```
