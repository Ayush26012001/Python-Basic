# Loop Statment
- It allows you to execute a block of code repeatedly.
- For Loop
- While Loop

## While Loop
- Repeatdely executed a block of code until a condition is met

```python
n=7
i=1
while i<n:
    print(i)
    i+=1
```

```python
count =5
while count>0:
    print(count)
    count-=1
```

```python
n=7
i=1
while i<n:
    print(i)
    i+=1
else:
    print("This will be executed when the while is run successfully without any break")
```

# Break Statment- It exists the loop

```python
n=7
i=1
while i<n:
    print(i)
    i+=1
    if i==3:
        break
else:
    print("This will be executed when the while is run successfully without any break")
```

# Continue Statment- It will skip the iteration

```python
n=7
i=1
while i<n:
    
    i+=1
    if i==3:
        continue
    print(i)
else:
    print("This will be executed when the while is run successfully without any break")
```

# For Loop
- Iterate over a sequence of element

```python
for i in "pwskills":
    print(i)
```

```python
a='ayush'
count=0
for i in a:
    count=count+1
print(count)
```

```python
a=12345
count=0
for i in str(a):
    count=count+1
print(count)
```

```python
a=[1,2,3,'a','b']
for i in a:
    print(i)
```

```python
a=[1,2,3,'a','b']
for i in a:
    print(i)
else:
    print("This is will print when for loop end without break")
```

```python
for i in a:
    if i=='a':
        break
    print(i)
else:
    print("This is will print when for loop end without break")
```

```python
for i in a:
    if i=='a':
        continue
    print(i)
else:
    print("This is will print when for loop end without break")
```

```python
row=1
while row<=4:
    col=1
    while col<=row:
        print("*",end="")
        col+=1
    print()
    row+=1
```

```python
for i in range(4):
    for j in range(i+1):
        print("*",end='')
    print()
```

```python
row=4
for i in range(row):
    for j in range(row-i):
        print("*", end='')
    print()
```

```python
row=4
for i in range(row):
    for j in range(row-i-1):
        print(" ",end="")
    for k in range(i+1):
        print("*",end='')
    print()
    
```

```python
row=3
for i in range(row):
    for j in range(row):
        print("*",end=" ")
    print()
```

```python
row=5
for i in range(row):
    print(" " * (row-i),end=' ')
    print("*" * (2 * i - 1))
```

```python
rows = 4

for i in range(rows, 0, -1):
    print(" " * (rows - i) + "*" * (2 * i - 1))
```

# Multiplication Table

```python
for i in range(1,6):
    for j in range(1,11):
        product=i*j
        print(f"{i} * {j} ={product}")
    print()
```

```python
r=3
for i in range(1,11):
    print(r,'*',i,"=",r*i)
```

```python
for i in range(5, 0, -1):
    print("*" * i)
```
